# 2D Laplace equation

Neural network for heat conduction problem with rectangular geometry

with $u(x) = y\sin (\pi x) + x(1-x)y(1-y)\cdot N(x)$

or $u(x) = y\sin (\pi x/2) + x(2-x)y(1-y)\cdot N(x)$

```julia
using Flux

# ------------------------------------------------------------
# training set
# ------------------------------------------------------------

x0 = 0.0; x1 = 2.0
y0 = 0.0; y1 = 1.0
nxTrain = 23; nyTrain = 11

xVec = Vector(range(x0, stop=x1, length=nxTrain))
#xVec = rand(nxTrain) .* (x1 - x0) .+ x0
xMesh = repeat(reshape(xVec, 1, :), nyTrain, 1)
yVec = Vector(range(y0, stop=y1, length=nyTrain))
#yVec = rand(nyTrain) .* (y1 - y0) .+ y0
yMesh = repeat(yVec, 1, nxTrain)

xMesh1D = reshape(xMesh, (1, :))
yMesh1D = reshape(yMesh, (1, :))
mesh = cat(xMesh1D, yMesh1D; dims=1)
X = deepcopy(mesh)

#X = rand(2, nxTrain * nyTrain)
Y = zeros(1, nxTrain * nyTrain)

m = Chain(Dense(2, 20, tanh), Dense(20, 20, tanh), Dense(20, 1))
```

```julia
mat_x(x) = [1.0 0.0] * x
mat_y(x) = [0.0 1.0] * x

u(x) = sin.(π .* mat_x(x) ./ 2.) .* mat_y(x) .+ 
    mat_x(x) .* (2. .- mat_x(x)) .* mat_y(x) .* (1. .- mat_y(x)) .* m(x)

ux(x) = mat_x(Tracker.forward(u, x)[2](1)[1])
uy(x) = mat_y(Tracker.forward(u, x)[2](1)[1])
uxx(x) =  mat_x(Tracker.forward(ux, x)[2](1)[1])
uyy(x) = mat_y(Tracker.forward(uy, x)[2](1)[1])

resi(x) = uxx(x) + uyy(x)
loss(x, y) = sum(resi(x).^2)
```

```julia
using Base.Iterators: repeated
using Flux: throttle
using Flux: @epochs

dataset = repeated((X, Y), 10)
evalcb = () -> @show(loss(X, Y))
opt = ADAM()
#opt = Nesterov()
ps = params(m)

#for (X,Y) in dataset
@epochs 5 for (X,Y) in dataset
    gs = Tracker.gradient(()->@show(loss(X,Y)), ps, nest=true)
    Tracker.update!(opt, ps, gs)
end
```

```julia
# ------------------------------------------------------------
# test set
# ------------------------------------------------------------

nxTest = 23; nyTest = 11

# meshgrid
xVecTest = Vector(range(x0, stop=x1, length=nxTest))
xMeshTest = repeat(reshape(xVecTest, 1, :), nyTest, 1)

yVecTest = Vector(range(y0, stop=y1, length=nyTest))
yMeshTest = repeat(yVecTest, 1, nxTest)

xMeshTest1D = reshape(xMeshTest, (1, :))
yMeshTest1D = reshape(yMeshTest, (1, :))
meshTest = cat(xMeshTest1D, yMeshTest1D; dims=1)

f_pred = u(meshTest)
f_pred = reshape(Tracker.data(f_pred), (nyTest, nxTest))
```

```julia
using Plots

p1 = contour(xVecTest, yVecTest, f_pred, fill=true)
plot(p1)
```
