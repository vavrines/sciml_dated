```julia
t0 = 0.f0; t1 = 3.f0; nt = 20
u0 = -4.f0; u1 = 4.f0; nu = 40;

t = range(t0, stop=t1, length=nt)
u = range(u0, stop=u1, length=nu)
tMesh, uMesh = meshgrid(t, u)

tMesh1D = reshape(tMesh, (1, :))
uMesh1D = reshape(uMesh, (1, :))
mesh = cat(tMesh1D, uMesh1D; dims=1)

X = deepcopy(mesh)
Y = zeros(Float32, 1, nt * nu);
```

```julia
inK = 2.f0
γ = (inK+3.f0) / (inK+1.f0)

f0 = zeros(Float32, nu);
f0 = u.^2 .* exp.(-u.^2)
weight = ones(Float32, nu) .* (u[nu] - u[nu-1]);

w0 = moments_conserve(f0, u, weight)
prim0 = conserve_prim(w0, γ)

g0 = maxwellian(u, prim0);
```

```julia
ann = FastChain(FastDense(2, 20, tanh), FastDense(20, 20, tanh), FastDense(20, 1))
p = initial_params(ann)

f(x, p) = mat_split(x)[2].^2 .* exp.(-mat_split(x)[2].^2) .+ mat_split(x)[1] .* ann(x, p)
```

```julia
function loss(θ)
     
    f(x) = mat_split(x)[2].^2 .* exp.(-mat_split(x)[2].^2) .+ mat_split(x)[1] .* ann(x, θ)
    
    #fx(x) = mat_split(pullback(f, x)[2](ones(size(x)))[1])[1]
    #uy(x) = pullback(u, x)[2](ones(size(x)))[1] |> mat_y
    
    #uxx(x) = pullback(ux, x)[2](ones(size(x)))[1] |> mat_x
    #uyy(x) = pullback(uy, x)[2](ones(size(x)))[1] |> mat_y
    
    function fx(x)
        ff = reshape(f(x), (nu, :))
        xx = reshape(X[1,:], (nu, :))
        dfx = zeros(eltype(x), nu, nt)
        for i in 1:nu
            dfx[i,:] .= central_diff(ff[i,:], xx[i,:])
        end
        
        return reshape(dfx, (1, :))
    end
    #=
    function uy(x)
        uu = reshape(u(x), (ny, :))
        yy = reshape(u0[2,:], (ny, :))
        duy = zeros(eltype(x), ny, nx)
        for j in 1:nx
            duy[:,j] .= central_diff(uu[:,j], yy[:,j])
        end
        
        return reshape(duy, (1, :))
    end
    #=
    function uxx(x)
        uu = reshape(ux(x), (ny, :))
        xx = reshape(u0[1,:], (ny, :))
        d2ux = zeros(eltype(x), ny, nx)
        for i in 1:ny
            d2ux[i,:] .= central_diff(uu[i,:], xx[i,:])
        end
        
        return reshape(d2ux, (1, :))
    end
    function uyy(x)
        uu = reshape(uy(x), (ny, :))
        yy = reshape(u0[2,:], (ny, :))
        d2uy = zeros(eltype(x), ny, nx)
        for j in 1:nx
            d2uy[:,j] .= central_diff(uu[:,j], yy[:,j])
        end
        
        return reshape(d2uy, (1, :))
    end
    
    #pred = uxx(u0) .+ uyy(u0)
    =#
    =#
    #dx = unstruct_diff(u, u0, ny, 1)
    #dy = unstruct_diff(u, u0, ny, 2)
    #d2x = unstruct_diff(ux, u0, ny, 1)
    #d2y = unstruct_diff(uy, u0, ny, 2)
        
    pred = fx(X) .+ f(X) .- prim0[1] .* (prim0[3] ./ π).^0.5 .* exp.(-prim0[3] .* (mat_split(X)[2] .- prim0[2]).^2)
    
    loss = sum(abs2, pred)
    
    return loss, pred
end
```

```julia
cb = function (θ, l, pred; isplot=false)
    display(l)
    return false
end
```

```julia
res = DiffEqFlux.sciml_train(loss, p, LBFGS(); cb=cb, maxiters=20, diffmode=DiffEqFlux.ReverseDiffMode())
```

```julia
for epoch in 1:10
    res = DiffEqFlux.sciml_train(loss, res.minimizer, LBFGS(); cb=cb, maxiters=500, diffmode=DiffEqFlux.ReverseDiffMode())
    res.minimum < 1 && break
end
```

```julia
function Kinetic.moments_conserve(f, u, ω)
    #w = zeros(3)
    #w[1] = discrete_moments(f, u, ω, 0)
    #w[2] = discrete_moments(f, u, ω, 1)
    #w[3] = 0.5 * discrete_moments(f, u, ω, 2)

    return [discrete_moments(f, u, ω, 0), discrete_moments(f, u, ω, 1), 0.5 * discrete_moments(f, u, ω, 2)]
end

Kinetic.discrete_moments(f,u,ω,n) = sum(@. ω * u^n * f)

function Kinetic.conserve_prim(W, γ)

    prim = zeros(eltype(W), axes(W))
    #prim = zeros(Float32, axes(W))

    prim[1] = W[1]
    prim[2] = W[2] / W[1]
    prim[3] = 0.5 * W[1] / (γ - 1.0) / (W[3] - 0.5 * W[2]^2 / W[1])

    return prim

end
```

```julia
function loss(θ)
     
    f(x) = mat_split(x)[2].^2 .* exp.(-mat_split(x)[2].^2) .+ mat_split(x)[1] .* ann(x, θ)
    
    function fx(x)
        ff = reshape(f(x), (nu, :))
        xx = reshape(X[1,:], (nu, :))
        dfx = zeros(eltype(x), nu, nt)
        for i in 1:nu
            dfx[i,:] .= central_diff(ff[i,:], xx[i,:])
        end
        
        return reshape(dfx, (1, :))
    end

    p1 = zeros(Float32, nu, nt)
    p2 = zeros(Float32, nu, nt)
    p3 = zeros(Float32, nu, nt)
    
    sol = reshape(f(X), nu, nt)
    for i in 1:nu, j in 1:nt
        W1 = discrete_moments(sol[:,j], u, weight, 0); 
        W2 = discrete_moments(sol[:,j], u, weight, 1); 
        W3 = 0.5 * discrete_moments(sol[:,j], u, weight, 2)
        p1[i,j] = W1
        p2[i,j] = W2 / W1
        p3[i,j] = 0.5 * W1 / (γ - 1.0) / (W3 - 0.5 * W2^2 / W1)
        
    end
    p1 = reshape(p1, (1,:))
    p2 = reshape(p2, (1,:))
    p3 = reshape(p3, (1,:))
        
    #p1, p2, p3 = prim0
    
    pred = fx(X) .+ f(X) .- p1 .* ((p3 ./ π).^2).^0.25f0 .* exp.(-p3 .* (mat_split(X)[2] .- p2).^2)
    loss = sum(abs2, pred)
    
    return loss, pred
end
```

```julia
for epoch in 1:1
    res = DiffEqFlux.sciml_train(loss, p, LBFGS(); cb=cb, maxiters=2, diffmode=DiffEqFlux.ReverseDiffMode())
    res.minimum < 1 && break
end
```
